step 2 > requests and CURL

```python
import requests

headers = {
    'sec-ch-ua': '"Google Chrome";v="117", "Not;A=Brand";v="8", "Chromium";v="117"',
    'sec-ch-ua-platform': '"Linux"',
    'Referer': 'https://www.ebooks.com/en-in/searchapp/searchresults.net?term=gore',
    'sec-ch-ua-mobile': '?0',
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36',
    'Content-Type': 'application/json',
}

params = {
    'term': 'gore',
    'CountryCode': 'IN',
}

response = requests.get('https://www.ebooks.com/api/search/', params=params, headers=headers)
```

step 7 - put everyting together - loop through results and append data inside a list 

```python
title =[]
subtitle=[]
author=[]
publisher=[]
py=[]
price=[]

for result in rj:
    title.append(result['title'])
    subtitle.append(result['subtitle'])
    author.append(result['authors'][0]['name'])
    publisher.append(result['publisher'])
    py.append(result['publication_year'])
    price.append(result['price'])
    
```

```python
price
```

step 8 > pandas dataframe

```python
bd = pd.DataFrame({'Title':title,'Subtitles':subtitle,'Author':author,'Publisher':publisher,'Publication_year':py,'Price':price})
```
